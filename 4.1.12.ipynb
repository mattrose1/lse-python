```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

movies = pd.read_csv('movies.csv')
ott = pd.read_excel('ott.xlsx')

print(movies.shape)
print(ott.shape)
print(movies.head())
print(ott.head())
print(movies.dtypes)
print(ott.dtypes)
```

```python
movies_ott = movies.merge(ott, on='ID', how='left')
```
