```python
import pandas as pd
import numpy as np

movies = pd.read_excel('movies_merge.xlsx')
ott = pd.read_csv('ott_merge.csv')

```

```python
movies_ott = movies.merge(ott, on='ID', how='left')
movies_ott
```

```python
movies_ott_runtime = movies_ott[['ID', 'Runtime', 'Genres']]

print(movies_ott_runtime)
```

```python
movies_ott_runtime['Runtime'].add(1)
```

```python
movies_ott_runtime['Gen_doc'] = np.where(movies_ott_runtime['Genres'].str.contains('Documentary'),'Documentary', 'Non-Documentary')

movies_ott_runtime
```
